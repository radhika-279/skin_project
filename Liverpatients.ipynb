This data set contains 415 liver patient records and 167 non liver patient records collected from North East of Andhra Pradesh, India. The "Target" column is a class label used to divide groups into liver patients (liver disease) or not (no disease). This data set contains 441 male patient records and 142 female patient records.

Here Target is Class, 1 : patient with  liver disease and 2: patient with no liver disease disease

## Explanation of attribute ##
Age: The age of the patient.

Gender: The gender of the patient, which can be either male or female.

Total Bilirubin: The total bilirubin level in the blood. Bilirubin is a substance produced during the normal breakdown of red blood cells and is processed by the liver.

Direct Bilirubin: The direct (conjugated) bilirubin level in the blood, representing the bilirubin that is processed by the liver and ready to be excreted from the body.

Alkaline Phosphotase: An enzyme related to the bile ducts; often increased when they are blocked.

Alamine Aminotransferase (ALT): An enzyme found mainly in the liver. High levels can indicate liver damage.

Aspartate Aminotransferase (AST): An enzyme found in several parts of the body, including the liver and muscles. Elevated levels can indicate liver damage or disease.

Total Proteins: The total amount of protein in the blood, which can give insights into the overall health of the liver.

Albumin: A protein made by the liver. Low levels can indicate liver disease or other health issues.

Albumin and Globulin Ratio: The ratio of albumin to globulin, which can be used to assess liver function. A low ratio can indicate liver disease.

Class: This is the class label used to indicate whether a patient has liver disease (1) or not (2).

Outliers present:
Total_Bilirubin,Direct_Bilirubin,Alkaline_Phosphotase,Alamine_Aminotransferase,Aspartate_Aminotransferase,Albumin_and_Globulin_Ratio

```python
outlier_columns=['Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase',
                 'Alamine_Aminotransferase','Aspartate_Aminotransferase','Albumin_and_Globulin_Ratio']

#handle oulier
def handle_outlies(df,column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound= Q1 - 1.5 * IQR
    upper_bound= Q3 + 1.5 * IQR
    df[column]=np.where(df[column] < lower_bound,lower_bound,df[column])
    df[column]=np.where(df[column] > upper_bound,upper_bound,df[column])

for col in outlier_columns:
    handle_outlies(df,col)

print("successfull")
                    
```

# XGBoost Model

```python
#split data by selected feature
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_res, y_train_res = smote.fit_resample(x_train, y_train)
x_train_res.shape , y_train_res.shape
```

# xgboost

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score,roc_curve, roc_auc_score, auc


xgb_model = xgb.XGBClassifier()
xgb_model.fit(x_train_res, y_train_res)

y_pred_xgb = xgb_model.predict(x_test)

# Evaluate the model
print("XGBoost Model Evaluation")
print(classification_report(y_test, y_pred_xgb))
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f'XGBoost Accuracy: {accuracy_xgb:.4f}')
f1_xgb = f1_score(y_test, y_pred_xgb)
print(f'XGBoost F1 score: {f1_xgb:.4f}')
```

```python
# Cross-validation and hyperparameter tuning with Stratified K-Fold
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV

xgb_model = xgb.XGBClassifier()

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.2, 0.3],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, 
                           cv=cv, n_jobs=-1, verbose=1, scoring='f1')

grid_search.fit(x_train, y_train)

print(f"Best Hyperparameters: {grid_search.best_params_}") # Best hyperparameters from GridSearchCV

best_xgb_model = grid_search.best_estimator_  # Use the best model from grid search for evaluation

y_pred_xgb = best_xgb_model.predict(x_test)

# Evaluate the model
print("XGBoost Model Evaluation")
print(classification_report(y_test, y_pred_xgb))
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f'XGBoost Accuracy: {accuracy_xgb:.4f}')
f1_xgb = f1_score(y_test, y_pred_xgb)
print(f'XGBoost F1 score: {f1_xgb:.4f}')

cm = confusion_matrix(y_test, y_pred_xgb)
print(f'confusion metrics:',cm)

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, best_xgb_model.predict_proba(x_test)[:,1])
roc_auc = auc(fpr, tpr)
print(f"XGBoost ROC AUC: {roc_auc:.4f}")
```
